```python
import mysql.connector as connection  # Here we create a table in mysql as in 'Salesclean'

try:
    mydb = connection.connect(host="localhost", database = 'clean',user="root", passwd='',use_pure=True)
    # check if the connection is established
    print(mydb.is_connected())

    query = "CREATE TABLE Salesclean ( product_code VARCHAR(60) ,customer_code VARCHAR(60)," \
            " markets_code VARCHAR(60), order_date DATE, sales_qty INT(10), sales_amount FLOAT, currency CHAR(3),"\
             "custmer_name VARCHAR(60), customer_type VARCHAR(60), markets_name VARCHAR(60), zone VARCHAR(60), product_type VARCHAR(60))"

    cursor = mydb.cursor() #create a cursor to execute queries
    cursor.execute(query)
    print("Table Created!!")
    mydb.close()
except Exception as e:
    mydb.close()
    print(str(e))
```
